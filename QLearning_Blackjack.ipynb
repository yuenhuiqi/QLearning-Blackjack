## 1. Simulator for drawing of cards & game between Player and Dealer

### 1.1 Simulating drawing of cards (Probability â€“ 0.75 Black, 0.25 Red)

Here, we're assigning 4 different ```col_num``` to each card. If the value of ```col_num``` is 1, 2 or 3, the card is Black. Otherwise, it's Red. This is to help simulate the probability of picking a Black card and the probability of picking a Red card.


````python
def draw_card():
    num = random.randint(1,10)
    col_num = random.randint(1,4)
    color = 'black' if col_num == 1 or col_num == 2 or col_num == 3  else 'red'
    return ([num, color])    
````

````python
def check_draw_card(n_sim=10000):
    count = 0
    for i in range(n_sim):
        card = draw_card()
        if card[1] == "black":
            count += 1
    print("Probability of drawing Black card after 10000 simulations: ", count/n_sim)
    print("Probability of drawing Red card after 10000 simulations: ", (n_sim - count)/n_sim)

check_draw_card()
````

We can see that in the ```check_draw_card()``` function, on average after 10,000 rounds of drawing a card, the overall probability of drawing a Black card is around 75% and the overall probability of drawing a Red card is around 25% which is the probability that we are looking for. As such, we'll be using the ```draw_card()``` function in the game to simulate the drawing of cards.

````python
# Function to calculate the total value of cards on hand
def total_val(hand):
    total = 0
    for card in hand:
        if card[1] == "red":
            total -= card[0]
        else: 
            total += card[0]
    return total

# Function to print out the hand and total value of both players at every decision epoch
def print_state(hand, p):
    print('\033[1m' + p + "'s hand:" + '\033[0m')
    for card in hand:
        print(str(card[0]) + ' ' + card[1]) 

# Function to decide outcome of game
def decide_outcome(player, dealer):
    player = total_val(player)
    dealer = total_val(dealer)
    
    if ((player > dealer) and (player < 22)) or ((player > 1) and (player <= 21) and (dealer > 21)):
        return 'PLAYER WON'
    elif (player == dealer) or ((player < 1 or player > 21) and (dealer < 1 or dealer > 21)):
        return 'DRAW'
    else:
        return 'DEALER WON'
````

### 1.2 Initializing Game

````python
def game():
    
    # First draw of cards before game starts: 
    player = [[random.randint(1, 10), 'black']]
    dealer = [[random.randint(1, 10), 'black']]
    
    print('\033[91m' + '\033[1m' + '\n****** GAME INITIALIZED ******' + '\033[0m')
    
    print("Player's starting card: ", str(player[0][0]), 'black')
    print("Dealer's starting card: ", str(dealer[0][0]), 'black')
    
    # Player's Turn
    print('\033[94m' + '\033[1m' + "\n****** PLAYER'S TURN ******" + '\033[0m')
    choice = input("Do you want to [H]it, [S]tick, or [Q]uit: ").lower()
    
    while choice != "q":
        if choice == "h" or choice == "hit":
            player.append(draw_card())
            print_state(player, 'Player')
            print('Current value: {}'.format(str(total_val(player))))
            
            # Bust
            if (total_val(player) < 1) or (total_val(player) > 21):
                print(total_val(player))
                print('\033[1m' + 'Player Busts\n' + '\033[0m')
                print('\033[91m' + '\033[1m' + '****** OUTCOME : ' + decide_outcome(player, dealer) +' ******\n' + '\033[0m')
                
                # Play again 
                again = input("Do you want to play again? (Y/N) : ").lower()
                if again == "y":
                    dealer = []
                    player = []
                    game()
                break
            else:
                choice = input("Do you want to [H]it, [S]tick, or [Q]uit: ").lower()
                
        
        elif choice == "s" or choice == "stick":
            print('\n' + '\033[95m' + '\033[1m' + "****** DEALER'S TURN ******" + '\033[0m')
            while total_val(dealer) < 16:
                print('\033[1m' + 'Dealer draws again\n' + '\033[0m')
                dealer.append(draw_card())
                print_state(dealer, 'Dealer')
                print('Current value: {}'.format(str(total_val(dealer))))
            
            # Win 
            if (total_val(dealer) >= 16) and (total_val(dealer) <= 21):
                print('\033[1m' + 'Dealer stops\n' + '\033[0m')
                print('\033[91m' + '\033[1m' + '****** OUTCOME : ' + decide_outcome(player, dealer) +' ******\n' + '\033[0m')
                # play again 
                again = input("Do you want to play again? (Y/N) : ").lower()
                if again == "y":
                    dealer = []
                    player = []
                    game()
                break

            # Bust
            elif (total_val(dealer) < 1) or (total_val(dealer) > 21):
                print('\033[1m' + 'Dealer busts\n' + '\033[0m')
                print('\033[91m' + '\033[1m' + '****** OUTCOME : ' + decide_outcome(player, dealer) +' ******\n' + '\033[0m')
                
                # Play again 
                again = input("Do you want to play again? (Y/N) : ").lower()
                if again == "y":
                    dealer = []
                    player = []
                    game()
                break
            
            print('\n') 
        else: 
            print("Please enter a valid input!\n")
            choice = input("Do you want to [H]it or [S]tick").lower()

    print("Goodbye!")
    exit()
        
# Instantiating Game
game()
````

## 2. Q-Learning Player
---

### 2.1 States & Actions for Q-Learning
<b>States:</b> The states here will be the 21 different scores that the player can be at when she decides whether to hit or stick. The score ranges from 1 to 21. Any score below 1 and above 21 will cause the player to bust, which means that the player has lost and isn't able to make the decision on whether to hit or stick. Hence, we'll be considering the scores 1 to 21 for this question. 

<b>Actions:</b> Hit or Stick. Action will be determined based on a non-deterministic strategy to allow for exploration of every arm. Here, we'll be using the Ð„-greedy bandit strategy to determine the action the Q-learning player would take. 

### 2.2 Initializing States & Actions as 0 in Q-table & Creating Ð„-greedy function

````python
alpha = 0.1
gamma = 0.9
epsilon = 0.1
episodes = 100000

Q_table = {}
actions = ['hit', 'stick']
for i in range(1, 22):
    for action in actions:
        Q_table[(i, action)] = 0
print(Q_table)

# Ð„-greedy function to determine what action to take
def epsilon_greedy(state):
    if random.random() < epsilon:
        action = getQ(state)[1][1]
    else:
        action = getQ(state)[0][1]
    return action
````

### 2.3 Functions to update & get Q values from Q-table

````python
def getQ(state):
    state_action = []
    for k, v in Q_table.items():
        if k[0] == state:
            
            # if there's no state actions, select the first state-action pair in Q-table that matches state.
            if len(state_action) == 0:
                state_action.append([k[0], k[1], v])
                
            # if there's existing state action â€“Â if value of another state action is higher, replace. 
            elif v > state_action[0][2]:
                state_action.insert(0, [k[0], k[1], v])
                
            # if the value of another state action is lower, just append to state_action list. 
            else:
                state_action.append([k[0], k[1], v])
                
    return state_action
    

def updateQ(old_state, action, new_state, reward):
    old_q = Q_table[(old_state, action)]
    
    try:
        new_q = getQ(new_state)[0][2]
        
    except IndexError:
        new_q = 0
        
     # Updating Q value of state-action pair
    Q_table[(old_state, action)] = old_q + alpha*(reward + gamma*new_q - old_q)
    
    
def saveQtable(self,file_name):  #save table
    with open(file_name, 'wb') as handle:
        pickle.dump(self.Q, handle, protocol=pickle.HIGHEST_PROTOCOL)

        
def loadQtable(self,file_name): # load table
    with open(file_name, 'rb') as handle:
        self.Q = pickle.load(handle)
````

````python
def train(num_iterations):
    for i in range(num_iterations):
        print("Training", i)
        player_val = random.randint(1, 10)
        dealer_val = random.randint(1, 10)
        
        # Training for Player
        action = 'hit'
        while action == 'hit':
            current_state = player_val
            action = epsilon_greedy(current_state)
            if action == 'hit':
                new_card = draw_card()
                if new_card[1] == 'red':
                    player_val -= new_card[0]
                else:
                    player_val += new_card[0]
            if player_val > 21 or player_val < 1:
                break
            else:
                # new state = new player's value, reward still 0 as game hasn't ended
                updateQ(current_state, action, player_val, 0)
        player_final_state = (current_state, action, player_val)

        # Dealer's moves
        action = 'hit'
        while action == 'hit': # repeat dealer's turn till stick or >=17 & <=21
            current = dealer_val
            if current >= 16 and current <= 21: # stick
                action = 'stick'
            else:
                action = 'hit'
                new_card = draw_card()
                if new_card[1] == 'black':
                    dealer_val += new_card[0]
                else:
                    dealer_val -= new_card[0]

                if dealer_val < 1 or dealer_val > 21: # bust
                    break
        dealer_final_state = (current, action, dealer_val)        
    
        # REWARDS
        
        # Outcome 1: Dealer Wins ( player < dealer (both didn't bust) OR player > 21 or < 1 AND dealer > 1 < 21 (player bust, dealer didn't bust))
        if ( (player_final_state[2] < dealer_final_state[2]) or ( (player_final_state[2] > 21 or player_final_state[2] < 1 ) and ( dealer_final_state[2] < 21 or dealer_final_state[2] > 1 ) ) ):
            updateQ(player_final_state[0], player_final_state[1], player_final_state[2], -1) # player's POV, player lose
            updateQ(dealer_final_state[0], dealer_final_state[1], dealer_final_state[2], 1) # dealer's POV, dealer win
            
        # Outcome 2: Player Wins ( player > dealer (both didn't bust) OR player < 21 or > 1 AND dealer < 1 > 21 (player didn't bust, dealer bust))
        elif ( (player_final_state[2] > dealer_final_state[2]) or ( (player_final_state[2] < 21 or player_final_state[2] > 1 ) and ( dealer_final_state[2] > 21 or dealer_final_state[2] < 1 ) ) ):
            updateQ(player_final_state[0], player_final_state[1], player_final_state[2], 1) # player's POV, player win
            updateQ(dealer_final_state[0], dealer_final_state[1], dealer_final_state[2], -1) # dealer's POV, dealer lose
        
        # Outcome 3: Draw
        else: 
            updateQ(player_final_state[0], player_final_state[1], player_final_state[2], 0) # player's POV, draw
            updateQ(dealer_final_state[0], dealer_final_state[1], dealer_final_state[2], 0) # dealer's POV, draw

# training for 100,000 iterations
train(episodes)
````
